# ASSIGNMENT 1
How to join two tables using DataFrame using different  types of join.

```python
import pandas as pd 
T = {"Student_ID":[1000,1001,1002,1003,1004],"totalmarks":[100,110,120,130,178],"ExtMarks":[80,75,65,90,30]}
T1 = pd.DataFrame(T,index=[1,2,3,4,5])
T1
```

```python
T2={"Student_ID":[1000,1001,1002,1003,1005],"Names":["Rossum","Gosling","Ritche","McKinney","Rooney"],"Subjects":["Python","Java","C","Pandas","ML"],"Ages":[65,80,85,55,45]}
T3=pd.DataFrame(T2,index=[1,2,3,4,5])
T3
```

# INNER JOIN

```python
merged_df1=pd.merge(T1,T3,on='Student_ID',how='inner')
merged_df1
```

# LEFT JOIN

```python
merged_df2=pd.merge(T1,T3,on='Student_ID',how='left')
merged_df2
```

# RIGHT JOIN

```python
merged_df3=pd.merge(T1,T3,on='Student_ID',how='right')
merged_df3
```

# OUTER JOIN

```python
merged_df4=pd.merge(T1,T3,on='Student_ID',how='outer')
merged_df4
```
